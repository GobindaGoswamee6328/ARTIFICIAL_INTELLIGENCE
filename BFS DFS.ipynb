# DFS using Adjacency List

```python
def readGraphData(file):
    f=open(file,"r")

    graph=[]

    for line in f.readlines():
        row=[]
        row.append(int(line.split()[0]))
        row.append(int(line.split()[1]))
        graph.append(row)
    return graph

graph=readGraphData("graph.txt")

print(graph)

vertex=graph[0][0]
edges=graph[0][1]
#print(vertex,edges)

adj={}

for i in range(vertex):
    adj[i]=[]

for e in range(1,len(graph),1):
    u=graph[e][0]
    v=graph[e][1]
    adj[u].append(v)
    adj[v].append(u)

    
for i in range(vertex):
    print(str(i)+"-> ",end=" ")
    print(adj[i])
    

print(adj)


import queue


def DFS_List(adj,s):
    stack = queue.LifoQueue()
    stack.put(s)

    visited=[]
    for j in range(len(adj)):
        visited.append(0)

    visited[s]=1
    while not stack.empty():
        node = stack.get()
        print(node,end=" ")
        for i in adj[node]:
            if visited[i]==0:
                stack.put(i)
                visited[i]=1
    
DFS_List(adj,0)
# 0 3 4 6 2 5 1
```

# Adjacency Matrix

```python
def readGraphData(file):
    f=open(file,"r")

    graph=[]

    for line in f.readlines():
        row=[]
        row.append(int(line.split()[0]))
        row.append(int(line.split()[1]))
        graph.append(row)
    return graph


def createMatrix(vertex):
    adj=[]
    for i in range(vertex):
        row=[]
        for j in range(vertex):
            row.append(0)
        adj.append(row)
    return adj

def printMatrix(M):
    for i in M:
        print(i,end="\n")


def addEdges(graphData,Matrix):
    for i in range(1,len(graphData)):
        edge=graphData[i]
        u=edge[0]
        v=edge[1]
        Matrix[u][v]=1
        #Matrix[v][u]=1
    return Matrix

import queue
def DFS_Matrix(Matrix,s):
    stack = queue.LifoQueue()
    stack.put(s)
    
    visited=[]
    for i in range(len(Matrix[0])):
        visited.append(0)
        
    visited[s]=1
    while not stack.empty():
        node = stack.get()
        print(node,end=" ")
        
        for j in range(len(Matrix[0])):
            if Matrix[node][j]==1 and visited[j]==0:
                stack.put(j)
                visited[j]=1
        
data = readGraphData("graph.txt")
print(data)

vertex=data[0][0]
edges=data[0][1]
#print(vertex,edges)
M=createMatrix(vertex)
#printMatrix(M)
Matrix = addEdges(data,M)
print()
printMatrix(Matrix)
DFS_Matrix(Matrix,0)
```
